```python
#import pandas library for reading data and merging
import pandas as pd
```

```python
#read demographic data
dem_data = pd.read_csv('Demograhic_Data.csv')
```

```python
dem_data.head()
```

```python
#read restaurants data
restaurant_data = pd.read_csv('Restaurants.csv')
```

```python
restaurant_data.head()
```

```python
#merge data using left join
combined_data = restaurant_data.merge(dem_data, left_on='city', right_on='City', how='left', copy=False)
```

```python
combined_data.head()
```

```python
#write final data file to csv
combined_data.to_csv('Final_Data.csv', index=False)
```
